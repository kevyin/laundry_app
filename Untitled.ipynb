```python
def open_weather_current(lat,lon,api='4fbe2c3c02d76185e9ebb7e2c24b54d0'):
    api_url = f"https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={api}&units=metric"
    response = requests.get(api_url)
    return response.json()

def open_weather_5day_3hr(lat,lon,api='4fbe2c3c02d76185e9ebb7e2c24b54d0'):
    api_url = f"https://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={api}&units=metric"
    print(api_url)
    response = requests.get(api_url)
    return response.json()

    
```

```python
def is_good_for_laundry(event):
    return (event['main']['humidity'] < 85) and ( event['clouds']['all'] < 10)

def to_here_date(dt_):
    from_zone = tz.tzutc()
    to_zone = tz.tzlocal()
    dt = datetime.utcfromtimestamp(dt_).replace(tzinfo=from_zone).astimezone(to_zone)
    return dt

def laundry_day_score(events):
    sum_cloud = 0
    sum_humidity = 0
    sum_isgood = 0
    for event in events:
        dt = to_here_date(event['dt'])
        
        is_good = is_good_for_laundry(event)
        humidity = event['main']['humidity']
        cloud = event['clouds']['all']
        print(dt, 'cloud', cloud)
        sum_humidity += humidity
        sum_cloud += cloud
        sum_isgood += 1 if is_good else 0
        
        # print(i,  dt.strftime('%A'), event['dt_txt'], event['weather'][0]['description'], is_good_for_laundry(event), humidity, cloud)
    avg_cloud = sum_cloud / len(events)
    avg_humidity = sum_humidity / len(events)
    avg_isgood = sum_isgood / len(events)
    return avg_cloud, avg_humidity, avg_isgood

def laundry_days(forecast_response):

    days = {}

    for i, event in enumerate(forcast_response['list']):
        dt = to_here_date(event['dt'])
        day = dt.strftime('%A %Y-%m-%d')

        # between 6am and 7
        if dt.hour >= 9 and dt.hour <= 19:
            if day in days:
                days[day].append(event)
            else:
                days[day] = [event]

    day_scores = []
    for day, day_events in days.items():
        print(day)
        cloud_score, humidity_score, isgood_score = laundry_day_score(day_events)
        print(day, cloud_score, humidity_score, isgood_score * 100)
        day_scores.append((day, cloud_score, humidity_score, isgood_score))

    # print(sorted(day_scores, key=lambda scores: scores[3]))
                
laundry_days(forcast_response)
    
# event = forcast_response['list'][0]
# # print(event)

# print(dt)
# for key, val in event.items():
#     print(key, val)
```
